# Regressão 01 - tarefa 02 - transformações em X e Y

### 1. Melhorando o ajuste no percentual de gorjetas

a. Vamos partir do modelo feito na última tarefa, o que relaciona ```tip_pct``` e ```net_bill```. Carregue a base, os pacotes e reproduza este modelo aqui.

````python
import pandas as pd
import seaborn as sns
import statsmodels.api as sm
import numpy as np
from sklearn.preprocessing import PolynomialFeatures
tips = sns.load_dataset('tips')
tips['tip_pct'] = tips['tip'] / tips['total_bill']
net_bill = tips['total_bill'] - tips['tip']
tip_pct = tips['tip_pct']
net_bill = sm.add_constant(net_bill)
modelo = sm.OLS(tip_pct, net_bill).fit()
modelo.summary()
````

### 2. ajuste o modelo de ```tip_pct``` em função do logaritmo de ```net_bill```.

````python
# Aplicando o logaritmo à variável net_bill
log_net_bill = np.log(tips['total_bill'] - tips['tip'])
# Adicionando uma constante à variável independente para a regressão
log_net_bill = sm.add_constant(log_net_bill)
# Realizando a regressão de tip_pct explicada pelo logaritmo de net_bill
modelo_log = sm.OLS(tip_pct, log_net_bill).fit()
modelo_log.summary()
````

###  3. ajuste o modelo de ```tip_pct``` em função de um polinômio de segundo grau em ```net_bill```.

````python
poly = PolynomialFeatures(degree=2)
X_poly = poly.fit_transform(tips['total_bill'].values.reshape(-1, 1))
# Realizando a regressão de tip_pct explicada pelo polinômio de segundo grau em net_bill
modelo_poly = sm.OLS(tip_pct, X_poly).fit()
modelo_poly.summary()
````

###  4. ajuste um modelo no log do ```tip_pct``` em função de ```net_bill```. Calcule o $R^2$ também em ```tip_pct```

````python
# Aplicando o logaritmo ao tip_pct
log_tip_pct = np.log(tip_pct)
# Adicionando uma constante à variável independente para a regressão
net_bill_with_const = sm.add_constant(net_bill)
# Realizando a regressão do logaritmo do tip_pct explicado por net_bill
modelo_log_tip_pct = sm.OLS(log_tip_pct, net_bill_with_const).fit()
# Calculando o R² para o modelo do logaritmo do tip_pct
R_squared_log_tip_pct = modelo_log_tip_pct.rsquared
# Calculando o R² para o modelo do tip_pct original
R_squared_tip_pct = 1 - (modelo_log_tip_pct.resid.var() / tip_pct.var())
print("R² para o modelo do logaritmo do tip_pct:", R_squared_log_tip_pct)
print("R² para o modelo do tip_pct original:", R_squared_tip_pct)
````

###  5. ajuste um modelo no log do ```tip_pct``` em função do log de ```net_bill```. Calcule o $R^2$ também em ```tip_pct```.

````python
log_net_bill = np.log(tips['total_bill'] - tips['tip'] + 1e-6)
# Aplicando o logaritmo ao tip_pct
log_tip_pct = np.log(tip_pct)
# Adicionando uma constante à variável independente para a regressão
log_net_bill_with_const = sm.add_constant(log_net_bill)
# Realizando a regressão do logaritmo do tip_pct explicado pelo logaritmo de net_bill
modelo_log_log = sm.OLS(log_tip_pct, log_net_bill_with_const).fit()
# Calculando o R² para o modelo do logaritmo do tip_pct
R_squared_log_log = modelo_log_log.rsquared
# Calculando o R² para o modelo do tip_pct original
residuals_tip_pct = np.exp(log_tip_pct) - np.exp(modelo_log_log.predict(log_net_bill_with_const))
R_squared_tip_pct = 1 - (residuals_tip_pct.var() / tip_pct.var())
print("R² para o modelo do logaritmo do tip_pct:", R_squared_log_log)
print("R² para o modelo do tip_pct original:", R_squared_tip_pct)
````

###  6. qual modelo resultou no melhor ajuste?

Comparando os valores de $R^2$ observamos que o modelo 4, ajustado com tip_pct em função do logaritmo de net_bill, apresentou o melhor ajuste com um $R^2$ de 0.2890. é o que resultou no melhor ajuste entre os modelos analisados.
