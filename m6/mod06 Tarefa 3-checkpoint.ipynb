#### 1) A base 'sinasc_RO_2019.csv' contém dados do sistema de nascidos vivos do DataSUS, cujo dicionário de dados está no arquivo 'Estrutura_SINASC_para_CD.pdf'. Carregue esse arquivo.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
sinasc = pd.read_csv('sinasc_RO_2019.csv')
```

#### 2) Categorize a variável APGAR5 conforme as seguintes quebras:

- 0 a 3: 'asfixia grave'
- 4 a 5: 'asfixia moderada'
- 6 a 7: 'asfixia leve'
- 8 a 10: 'normal'

```python
def categorizar_apgar5(apgar5):
    if apgar5 in range(0, 4):
        return 'asfixia grave'
    elif apgar5 in range(4, 6):
        return 'asfixia moderada'
    elif apgar5 in range(6, 8):
        return 'asfixia leve'
    else:
        return 'normal'

sinasc['APGAR5_Categoria'] = sinasc['APGAR5'].apply(categorizar_apgar5)
```

#### 3) Monte um gráfico comparando a distribuição de APGAR5 por sexo para cada grupo de tempo de gestação (GESTACAO).

```python
plt.figure(figsize=(12, 8))
sns.countplot(data=sinasc, x='GESTACAO', hue='APGAR5_Categoria', palette='viridis')
plt.title('Distribuição de APGAR5 por sexo para cada grupo de tempo de gestação')
plt.xlabel('Tempo de Gestação')
plt.ylabel('Contagem')
plt.legend(title='APGAR5 Categoria')
plt.xticks(rotation=45)
plt.show()
```

####  4) Monte uma tabela cruzada com APGAR5 nas colunas e GESTACAO nas linhas

```python
tabela_cruzada = pd.crosstab(sinasc['GESTACAO'], sinasc['APGAR5_Categoria'])
```

####  5) Da tabela cruzada do intem anterior, calcule os percentuais de linha. Ou seja, a distribuição da variável APGAR5 para cada classe de GESTACAO dado.

```python
percentuais_linha = tabela_cruzada.div(tabela_cruzada.sum(axis=1), axis=0)
```

####  6) Os médicos costumam classificar parto prematuro quando a gestação tem menos que 37 semanas. Faça um gráfico de barras da tabela cruzada acima e conclua se você acha que este conjunto de dados corrobora ou não com a afirmação dos médicos.

```python
plt.figure(figsize=(12, 8))
percentuais_linha.plot(kind='bar', stacked=True, cmap='viridis')
plt.title('Distribuição de APGAR5 por tempo de gestação (Percentuais de linha)')
plt.xlabel('Tempo de Gestação')
plt.ylabel('Percentual')
plt.legend(title='APGAR5 Categoria')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
