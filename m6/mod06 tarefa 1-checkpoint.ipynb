#### 1) A base 'sinasc_RO_2019.csv' contém dados do sistema de nascidos vivos do DataSUS, cujo dicionário de dados está no arquivo 'Estrutura_SINASC_para_CD.pdf'. Carregue esses arquivos.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
sinasc = pd.read_csv('SINASC_RO_2019.csv')
```

####  2) Monte um gráficos de barras com as variáveis APGAR1, APGAR5, SEXO, GESTACAO, ESTCIVMAE.

```python
gestacao_order = ['Menos de 22 semanas', '22 a 27 semanas', '28 a 31 semanas', '32 a 36 semanas', '37 a 41 semanas', '42 semanas e mais']
fig, axs = plt.subplots(3, 2, figsize=(14, 12))
sinasc['APGAR1'].value_counts().sort_index().plot(kind='bar', ax=axs[0, 0], title='APGAR1')
sinasc['APGAR5'].value_counts().sort_index().plot(kind='bar', ax=axs[0, 1], title='APGAR5')
sinasc['SEXO'].value_counts().plot(kind='bar', ax=axs[1, 0], title='SEXO')
sinasc['GESTACAO'].value_counts().loc[gestacao_order].plot(kind='bar', ax=axs[1, 1], title='GESTACAO')
sinasc['ESTCIVMAE'].value_counts().plot(kind='bar', ax=axs[2, 0], title='ESTCIVMAE')
fig.delaxes(axs[2, 1])
plt.tight_layout()
plt.show()
```

#### 3) Monte um gráfico de barras mostrando o APGAR5 médio por CONSULTAS

```python
apg5_media_por_consultas = sinasc.groupby('CONSULTAS')['APGAR5'].mean()
apg5_media_por_consultas.plot(kind='bar', figsize=(8, 6), color='skyblue')
plt.title('APGAR5 Médio por Consultas')
plt.xlabel('Consultas')
plt.ylabel('APGAR5 Médio')
plt.xticks(rotation=45)
plt.show()
```

####  4) Monte um painel com 4 gráficos mostrando as frequencias de SEXO, GESTACAO, ESTCIVMAE e CONSULTAS

```python
fig, axs = plt.subplots(2, 2, figsize=(12, 10))

sinasc['SEXO'].value_counts().plot(kind='bar', ax=axs[0, 0], title='Frequência de SEXO')
sinasc['GESTACAO'].value_counts().plot(kind='bar', ax=axs[0, 1], title='Frequência de GESTACAO')
sinasc['ESTCIVMAE'].value_counts().plot(kind='bar', ax=axs[1, 0], title='Frequência de ESTCIVMAE')
sinasc['CONSULTAS'].value_counts().plot(kind='bar', ax=axs[1, 1], title='Frequência de CONSULTAS')

plt.tight_layout()
plt.show()
```
