```python
import pandas as pd
import matplotlib.pyplot as plt
sinasc = pd.read_csv('SINASC_RO_2019.csv')
```

####  1) Na base do SINASC do DataSUS, faça um boxplot da idade da mãe.

```python
plt.figure(figsize=(8, 6))
plt.boxplot(sinasc['IDADEMAE'].dropna(), vert=False)
plt.title('Boxplot da Idade da Mãe')
plt.xlabel('Idade')
plt.show()
```

####  2) Faça um boxplot do peso do bebê

```python
plt.figure(figsize=(8, 6))
plt.boxplot(sinasc['PESO'].dropna(), vert=False)
plt.title('Boxplot do Peso do Bebê')
plt.xlabel('Peso (g)')
plt.show()
```

####  3) Faça um histograma da idade da mãe

```python
plt.figure(figsize=(8, 6))
plt.hist(sinasc['IDADEMAE'].dropna(), bins=20, color='skyblue', edgecolor='black')
plt.title('Histograma da Idade da Mãe')
plt.xlabel('Idade')
plt.ylabel('Frequência')
plt.grid(True)
plt.show()
```

#### 4) Faça um histograma do peso do bebê

```python
plt.figure(figsize=(8, 6))
plt.hist(sinasc['PESO'].dropna(), bins=20, color='lightgreen', edgecolor='black')
plt.title('Histograma do Peso do Bebê')
plt.xlabel('Peso (g)')
plt.ylabel('Frequência')
plt.grid(True)
plt.show()
```
