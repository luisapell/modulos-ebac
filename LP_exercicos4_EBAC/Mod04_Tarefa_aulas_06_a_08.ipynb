####  2) Melhorando a interação com o API
Observe dois URLs de consultas diferentes, por exemplo o URL utilizado em aula, e o URL feito no exercício anterior. Compare-os e observe as diferenças.

1. Faça uma função em Python que recebe como argumento o UF da consulta e o tipo de interessado (```'Estado'```ou ```Município```), e que devolve os dados da consulta no formato *DataFrame*.
2. Quantas solicitações para o Estado podem ser consultadas para Minas Gerais com *status* em 'Arquivado por decurso de prazo' estão registradas?
3. Qual é o município da Bahia com mais solicitações deferidas?
4. Salve um arquivo .csv com os dados de solicitações da Bahia, com interessado = 'Estado'

````python
#1) Seu código aqui
def consulta_dados(uf, tipo_interessado):
    # Filtra os dados com base no UF e tipo de interessado
    dados_filtrados = df[(df['uf'] == uf) & (df['tipo_interessado'] == tipo_interessado)]
    return dados_filtrados
````

````python
# 2) Seu código aqui
uf_mg = 'MG'
tipo_interessado_mg = 'Estado'
resultado_tipo_interessado = consulta_dados(uf_mg, tipo_interessado_mg)
resultado_tipo_interessado['status'].value_counts()
status_MG = resultado_tipo_interessado['status'] == 'Arquivado por decurso de prazo'
status_MG.value_counts()
````

````python
uf_mg = 'MG'
tipo_interessado_consulta = 'Estado'
resultado_tipo_interessado = consulta_dados(uf_mg, tipo_interessado_mg)

# Contar o número de solicitações com status 'Arquivado por decurso de prazo'
qtd_solicitacoes_arquivadas = (resultado_tipo_interessado['status'] == 'Arquivado por decurso de prazo').sum()

print(f"Para o Estado de Minas Gerais com status 'Arquivado por decurso de prazo' é: {qtd_solicitacoes_arquivadas}")
````

````python
#3)Qual é o município da Bahia com mais solicitações deferidas?
uf_ba = 'BA'
tipo_interessado_ba = 'Município'
resultado_consulta = consulta_dados(uf_ba, tipo_interessado_ba)
deferido_data = resultado_consulta[resultado_consulta['status'] == 'Deferido']
deferido_ba = deferido_data['interessado'].value_counts()

print("O município da Bahia com mais solicitações deferidas é:", deferido_ba.idxmax())
````

````python
#4) Salve um arquivo .csv com os dados de solicitações da Bahia, com interessado = 'Estado'
dados_bahia_estado = df[(df['uf'] == 'BA') & (df['tipo_interessado'] == 'Estado')]
dados_bahia_estado.to_csv("solicitacoes_bahia_estado.csv", index=False)

print("Arquivo 'solicitacoes_bahia_estado.csv' salvo com sucesso.")
````
